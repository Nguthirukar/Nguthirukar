### Visualing the Correlation of the Dataset to understand key areas of insights

*   Import necessary libraries
*   Plot a heatmap to visualize the correlation matrix



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr = df.corr()
plt.figure(figsize = (15, 10))
sns.heatmap (corr, annot = True)
```

### Data Visualition of Correlations based on the Correlation matrix heatmap above using Scatter Plot

*   Relationship between Unit cost and Unit Price
*   Relationship between Revenue and Profits
-   Relationship between Revenue and Cost over Time(years)



Corr plot of Unit cost over Unit price

```python
df.plot(kind='scatter',x="Unit Cost", y="Unit Price",figsize=(12,4),fontsize=10)
plt.title("Relation between Unit cost & Unit Price",fontsize=17)
plt.ylabel("Unit Price",fontsize=15)
plt.xlabel("Unit Cost",fontsize=15)
```

Corr plot between Revenue and Profit

```python
df.plot(kind='scatter', x='Revenue', y='Profit', figsize=(10,8),fontsize=12)
plt.title("Relation between Revenue & Profit",fontsize=17)
plt.xlabel("Revenue",fontsize=13)
plt.ylabel("Profit",fontsize=13)
```

```python
# Correlation plot Revenue vs Cost

sns.scatterplot(data=df, x="Revenue", y="Cost", hue='year')
plt.xlim(0,3000)
plt.ylim(0,2000)
plt.gcf().set_size_inches(12,6)
plt.title("Relation between Revenue & Cost over time",fontsize=17)

m, b = np.polyfit(df["Revenue"], df["Cost"], 1)

plt.plot(df["Revenue"], m*df["Revenue"] + b)
```
